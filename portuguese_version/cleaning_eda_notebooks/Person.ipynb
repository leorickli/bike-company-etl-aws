```python
import pandas as pd

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('Person.Person.csv', sep=';', encoding='utf-8')

# Applying a lambda function to each element in the DataFrame 'df'.
# The lambda function checks if the element is NaN (Not a Number).
# If it is NaN, the element remains unchanged; otherwise, it encodes the element into ASCII and then decodes it back to remove any non-ASCII characters.
df = df.applymap(lambda x: x if pd.isna(x) else str(x).encode('ascii', 'ignore').decode('ascii'))
df.head()
```

```python
df.to_csv('Person.csv', index=False)
```
