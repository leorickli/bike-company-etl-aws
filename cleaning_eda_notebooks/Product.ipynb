```python
import pandas as pd

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None) 
```

```python
df = pd.read_csv('Production.Product.csv', sep=';')
df.head()
```

```python
# Remove all "null" cases
df = df.replace(r'(?i)NULL', '', regex=True)

# Substitute commas for dots
df['StandardCost'] = df['StandardCost'].str.replace(',', '.')
df['ListPrice'] = df['ListPrice'].str.replace(',', '.')

df.head()
```

```python
df.to_csv('Product.csv', index=False)
```
