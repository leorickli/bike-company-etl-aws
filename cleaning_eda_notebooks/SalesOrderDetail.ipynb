```python
import pandas as pd

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None) 
```

```python
df = pd.read_csv('Sales.SalesOrderDetail.csv', sep=';')
df.head()
```

```python
df.dtypes
```

```python
# Remove all "null" cases in upper and lower case
df = df.replace(r'(?i)NULL', '', regex=True)

# Substitute commas for dots
df['UnitPrice'] = df['UnitPrice'].str.replace(',', '.')
df['UnitPriceDiscount'] = df['UnitPriceDiscount'].str.replace(',', '.')

df.head()
```

```python
df.to_csv('SalesOrderDetail.csv', index=False)
```
