```python
import pandas as pd

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None) 
```

```python
df = pd.read_csv('Sales.Customer.csv', sep=';')
df.rename(columns=lambda x: x.strip(), inplace=True)

# Clean the 'StoreID' column by stripping leading and trailing white spaces
df['StoreID'] = df['StoreID'].str.strip()

# Convert the 'StoreID' column to integers, replacing empty values with NaN
df['StoreID'] = pd.to_numeric(df['StoreID'], errors='coerce', downcast='integer')

# Drop rows with NaN values in the 'StoreID' column
df.dropna(subset=['StoreID'], inplace=True)
df.head()
```

```python
# Remove all "null" cases
df = df.replace(r'(?i)NULL', '', regex=True)

df.head()
```

```python
df.to_csv('Customer.csv', index=False)
```
