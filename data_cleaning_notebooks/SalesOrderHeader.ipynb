```python
import pandas as pd

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None) 
```

```python
df = pd.read_csv('Sales.SalesOrderHeader.csv', sep=';')

df.head()
```

```python
# Remover todos os casos em que há "null"
df = df.replace(r'(?i)NULL', '', regex=True)

# Convert the 'StoreID' column to integers, replacing empty values with NaN
df['SalesPersonID'] = pd.to_numeric(df['SalesPersonID'], errors='coerce', downcast='integer')
# Drop rows with NaN values in the 'StoreID' column
df.dropna(subset=['SalesPersonID'], inplace=True)

# Convert the 'StoreID' column to integers, replacing empty values with NaN
df['CreditCardID'] = pd.to_numeric(df['CreditCardID'], errors='coerce', downcast='integer')
# Drop rows with NaN values in the 'StoreID' column
df.dropna(subset=['CreditCardID'], inplace=True)

# Substitute commas for dots
df['SubTotal'] = df['SubTotal'].str.replace(',', '.')
df['TaxAmt'] = df['TaxAmt'].str.replace(',', '.')
df['Freight'] = df['Freight'].str.replace(',', '.')
df['TotalDue'] = df['TotalDue'].str.replace(',', '.')

df.head()
```

```python
df.to_csv('SalesOrderHeader.csv', index=False)
```
